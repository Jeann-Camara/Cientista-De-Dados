## 1. Existe alguma relação entre a concentração de CO ao decorrer do dia?

```python
# Renomeia Colunas
df_Air_Quality_UCI = df_Air_Quality_UCI.rename(columns = {"Time": "Hora", "CO(GT)":"Concentração de CO (mg/m³)"}) 
```

```python
# Cria gráfico de barras com relação entre "Hora" e "CO".
plt.figure(figsize=(16, 6))
seb.barplot(data=df_Air_Quality_UCI, x="Hora", y="Concentração de CO (mg/m³)", estimator=np.mean)
```

## Portanto, existe relação entre a concentração de CO durante o dia, cujos "horários de pico", 08:00 às 10:00 e 18:00 às 20:00, repercutem em mais concentração de CO.

## 2.Existe alguma relação entre a concentração de CO entre os dias da semana?

```python
# Cria a coluna "Dia da Semana" apartir da coluna "Date" para a análise.
df_Air_Quality_UCI["weekday"] = pd.to_datetime( df_Air_Quality_UCI["Date"]).dt.weekday_name

# Cria função para traduzir Dias da Semana.
def Traduz_Dia_Semana(weekday_name):
    switcher = {
        
        "Monday": "1.Segunda",
        "Tuesday": "2.Terça",
        "Wednesday": "3.Quarta",
        "Thursday": "4.Quinta",
        "Friday":  "5.Sexta",
        "Saturday": "6.Sábado",
        "Sunday": "7.Domingo"
    }
    return switcher.get(weekday_name)

# Traduz dias da Semana
df_Air_Quality_UCI["weekday"] = df_Air_Quality_UCI["weekday"].apply(lambda x: Traduz_Dia_Semana(x))

# Ordena pelo dia da semana.
df_Air_Quality_UCI.sort_values(by=["weekday"], inplace=True)

# Trata ordenação do dia da semana.
df_Air_Quality_UCI["Dia da Semana"] = df_Air_Quality_UCI["weekday"].apply(lambda x: x[2:])
```

```python
# Cria gráfico de barras com relação entre "Dia da Semana" e "CO".
plt.figure(figsize=(16, 6))
seb.barplot(data=df_Air_Quality_UCI, x="Dia da Semana", y="Concentração de CO (mg/m³)", estimator=np.mean)
```

## Portanto, existe relação entre a concentração de CO e os dias da Semana, dos quais o fim de semana tem concentração menor de CO.

## 4. Predição utilizando o modelo KNN.

```python
# Verifica qual a correlação absoluta mais forte.
df_Air_Quality_UCI.corr()[abs(df_Air_Quality_UCI.corr()) < 1].max().sort_values(ascending=False).nlargest(n=2, keep="first")
```

```python
# Cria gráfico de barras com relação entre os atributos com maior correlação.
plt.figure(figsize=(16, 6))
plt.xlabel("PT08.S2(NMHC)")
plt.ylabel("C6H6(GT)")
plt.scatter(data=df_Air_Quality_UCI, x="PT08.S2(NMHC)", y="C6H6(GT)")
plt.show()
```

## Baseado na correlação absoluta mais forte, arbitro o "Titânio-NMHC/h" como variável independente e o "benzeno (mg/m³)" como variável alvo da predição.

```python
# Seleciona colunas para análise.
df_Titanio_Benzeno = df_Air_Quality_UCI[["PT08.S2(NMHC)","C6H6(GT)"]]
```

```python
# Renomeia Colunas
df_Titanio_Benzeno = df_Titanio_Benzeno.rename(columns = {"PT08.S2(NMHC)":"Titânio-NMHC/h", "C6H6(GT)": "benzeno(mg/m³)"}) 
```

```python
# Separa a coluna independente e a dependente para o modelo de predição.
x = df_Titanio_Benzeno["Titânio-NMHC/h"]
y = df_Titanio_Benzeno["benzeno(mg/m³)"]
```

```python
# Normalização dos dados.
x =  (x-min(x)) / (max(x) - min(x) )
```

```python
# Divide o dataset entre treino e teste. (70% treino e 30% teste) - Evitar OverFitting.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
# Instancia modelo KNN do sklearn.
knn_Regressor = KNeighborsRegressor(n_neighbors=1)
```

```python
# Treina o modelo.
knn_Regressor.fit(x_train.values.reshape(-1,1), y_train)
```

```python
# Utiliza modelo treinado para a predição.
y_pred = knn_Regressor.predict(x_test.values.reshape(-1,1))
```

```python
print("Acurácia KNN: %.1f %%" % (metrics.r2_score(y_test, y_pred)*100))
```

## A acurácia do modelo KNN para Regressão é em torno de 99%.

## 5.Predição utilizando o modelo de Regressão Linear.

```python
# Instancia modelo de Regressão Linear do sklearn.
reg_lin = linear_model.LinearRegression()
```

```python
# Treina o modelo.
reg_lin.fit(x_train.values.reshape(-1,1), y_train)
```

```python
# Utiliza modelo treinado para a predição.
y_pred = reg_lin.predict(x_test.values.reshape(-1,1))
```

```python
print("Acurácia Regressão Linear: %.1f %%" % (metrics.r2_score(y_test, y_pred)*100))
```

## A Acurácia do modelo de Regressão Linear para os mesmos dados de treino e teste é em torno de 96%, portanto, menor que o de KNN.
