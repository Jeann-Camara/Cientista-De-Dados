```r
# Especificando os parmetros

# col - cor do plot
# lty - tipo de linha
# lwd - largura de linha
# pch - smbolo no plot
# xlab - label do eixo x
# ylab - label do eixo y
# las - como os labels dos eixos so orientados
# bg - background color
# mfrow - nmero de plots por linha
# mfcool - nmero de plots por coluna

# Funes Bsicas de Plot

# plot() - scatterplots
# lines() -  adiciona linhas ao grfico
# points() - adiciona pontos ao grfico
# text() - adiciona label ao grfico
# title() - adiciona ttulo ao grfico
```

```r
# Parmetros dos Grficos
?par
par()
par('pch')
par('lty')

x = 2:4
plot(x, pch = "*")

par(mfrow = c(2,2), col.axis = "red")
plot(1:8, las = 0, xlab = "xlab", ylab = "ylab", main = "LAS = 0")
plot(1:8, las = 1, xlab = "xlab", ylab = "ylab", main = "LAS = 1")
plot(1:8, las = 2, xlab = "xlab", ylab = "ylab", main = "LAS = 2")
plot(1:8, las = 3, xlab = "xlab", ylab = "ylab", main = "LAS = 3")
legend("topright", pch = 1, col = c("blue", "red"), legend = c("Var1","Var2"))
par(mfrow = c(1,1))
```

```r
# Funo para a escala
mycex <- function(var, r, f = sqrt){
  x = f(var)
  x_scaled = (x - min(x))/(max(x) - min(x))
  r[1] + x_scaled * (r[2] - r[1])
}
```

```r
# Plot
plot(expectativa ~ pibpercap, data = df_1982, log = "x",
     col = mycol[continente],
     cex = mycex(pop, r = c(0.2, 10))
)
```
