## Análise 2

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
from datetime import datetime
sea.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Número de veículos pertencentes a cada marca

```python
# Plot com o número de veículos pertencentes a cada marca
sea_factor = sea.factorplot(x="brand", data=df, kind="count")
```

```python
# Salvando o plot
sea_factor.fig.savefig("plots/Analise1/count-vehicleType.png")
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Plot com o Preço médio dos veículos com base no tipo de veículo e no tipo de caixa de câmbio
sea_barplot = sea.barplot(x="vehicleType", y="price",hue="gearbox", data=df)
```

```python
# Salvando o plot
sea_barplot.get_figure().savefig("plots/Analise2/vehicletype-gearbox-price.png")
```
